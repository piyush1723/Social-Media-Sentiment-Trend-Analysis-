```python
import re
import pandas as pd


RAW_DATA_PATH = "../data/raw/twitter_training.csv"
PROCESSED_DATA_PATH = "../data/processed/clean_twitter_training.csv"
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"@\w+", " ", text)
    text = re.sub(r"http\S+|www\S+", " ", text)
    text = re.sub(r"#", "", text)
    text = re.sub(r"[^a-zA-Z\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python

df = pd.read_csv(
    RAW_DATA_PATH,
    header=None,
    names=["tweet_id", "entity", "sentiment", "tweet"],
)
```

STOPWORDS REMOVING

```python
df = df.dropna(subset=["tweet", "sentiment"])
df["clean_tweet"] = df["tweet"].apply(clean_text)
df = df[df["clean_tweet"].str.len() > 0]
```

```python
df.to_csv(PROCESSED_DATA_PATH, index=False)
```
